### Pre trained model from Google news text data

- https://stackoverflow.com/questions/46433778/import-googlenews-vectors-negative300-bin

- Input: 3Billion words from google news
- Output: 300Dimensions that is window size

```python
import gensim.downloader as api

wv= api.load("word2vec-google-news-300")
#300 here is window size

vec_king= wv["king"]
```

```python
vec_king #vector of King has 300 dim
```

```python
wv['man'] #vector of man
```

```python
# Similarities : most_similar()

wv.most_similar("man")
#this shows cosine similarity 
# also it gives synonyms
```

```python
wv.most_similar("dog")
```

```python
wv.most_similar("Melissa")
```

```python
wv.most_similar("Whisky")
```

```python
wv.most_similar("Sourav")
#Sourav: similar words are Ganguly and then cricket players
```

```python
#it returns synonyms and antonyms as well 

wv.most_similar("happy")
```

```python
# to find the similarity between words
#most_similar takes 1 argument and similarity() takes 2 words 

wv.similarity("King", "Man")
#there is similarity genderwise
```

```python
wv.similarity("King", "Queen")
#royalty wise there is similarity 
```

```python
wv.similarity("Sourav", "Melissa")
# there is no much similarity 0.011
```

```python
wv.similarity("Whisky", "Dog")
```

```python
vec1= wv['king']-wv['man']
wv.most_similar([vec1])

#just printing vec1 will return 300 dimension vector
```

```python
vec2= wv['king']-wv['man']+wv['woman']
wv.most_similar([vec2])
```
