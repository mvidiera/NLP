```python
#How to use NLTK

import nltk
#to apply stemming
from nltk.stem import PorterStemmer
#to apply stopwords
from nltk.corpus import stopwords
```

### Clening corpus 

```python
len(sentences)
```

```python
import re
corpus=[] #new corpus after cleaning
for i in range(len(sentences)):
    review= re.sub('[^a-zA-Z]',' ', sentences[i]) # substiture/ replace special character to blank
    review= review.lower() #converting into lower case
    #review= review.split()
    corpus.append(review)
```

```python
corpus
# Every words are changed into lower case and other than letter like special characters are removed 
#and replaced by spaces
```

```python
stopwords.words("english")
```

```python
#applying stemming on corpus

for i in corpus: #i has sentences
    words= nltk.word_tokenize(i) # from sentence Im getting words
    
    for word in words: # in i(sentences) for each word
        if word not in set(stopwords.words("english")): # for all words which are not stop words
            print(stemmer.stem(word))
```

```python
for i in corpus: 
    words= nltk.word_tokenize(i) 
    
    for word in words: 
        if word not in set(stopwords.words("english")):
            print(lemmatizer.lemmatize(word))
```

### Bag of Words

```python
#count vectorisation
# I copied this from google: class sklearn.feature_extraction.text.CountVectorizer
# to use count vectorization: 
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv= CountVectorizer()
```

```python
#The fit(data) method is used to compute the mean and std dev for a given feature to be used 
#further for scaling. 
#The transform(data) method is used to perform scaling using mean and std dev calculated using 
#the . fit() method. The fit_transform() method does both fits and transform.

X= cv.fit_transform(corpus)
```

```python
X
```

```python
# to make it binary BoW, change binary parameter in count vectorizer
# taking new obj 

cv1= CountVectorizer(binary= True)
X= cv1.fit_transform(corpus)
X[0].toarray()

#now the BoW have turned into binary (other than 0 and 1, no other values)
```

### Applying Stopwords

```python
#lemmatization

import re
corpus1= []

for i in range(len(sentences)):
    review= re.sub('[^a-zA-Z]',' ', sentences[i])
    review= review.lower()
    review= review.split()
    review= [lemmatizer.lemmatize(word) for word in review if not word in set(stopwords.words("english"))]
    review= ' '.join(review)
    corpus1.append(review)
```

```python
#taking new object

cv3= CountVectorizer()
X1= cv3.fit_transform(corpus1)
```

```python
cv3.vocabulary_

#here stop words like the, of, was, is words are removed.
```

### TF-IDF

##### Disdvantages of BoW
1. It doesnt provide weightage to important or rare words within the sentences
2. Though sentences are opposite (ex: Not good, Good) BoW suggests that the sentences are close or similar.
3. Sematics or relationship between words are not captured

```python
#class sklearn.feature_extraction.text.TfidfVectorizer

from sklearn.feature_extraction.text import TfidfVectorizer
cv6= TfidfVectorizer()
```

```python
corpus1[0]# corpus1 has stopwords removed
```

```python
X6= cv6.fit_transform(corpus1)
```

```python
X6[0].toarray()

# vectors getting created has different values for each words. Ranges between 0 and 1
#0 being not so important words
# Higher the value (closer to 1) greater importance the word is provided 
```

```python
#Ngram in TFIDF 

cv7= TfidfVectorizer(ngram_range=(2,3))
X7= cv7.fit_transform(corpus1)
X7[0].toarray()

# noe values will be different as last code has 1 gram, values were given for each word
#here I have 2,3 grams which are 2gram followed by 3 gram 
```

#### max feature in TFIDF

- Max_feature is the parameter, if I mention this parameter like max_feature=3, then the words which are TOP3 according to term fequency(TF), only those words are considered and TFIDF matrix is created.
